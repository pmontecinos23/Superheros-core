## Superheros

```python
# Load user data
df = pd.read_csv(r"C:\Users\paulm\Documents\Saved Data for DataFrames - Coding Dojo\superhero_info - superhero_info.csv")
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
## make columns lowercase for ease of reference and display names
df.columns = df.columns.str.lower()
df.columns
```

```python
df.describe(include='all')
```

```python
# create 2 new columns, FirstName and LastName by splitting the Name column
df[['hero','publisher']] = df['hero|publisher'].str.split('|', expand=True)
# drop the 'Name' column
df.drop('hero|publisher', axis=1, inplace=True)
df.head()
```

```python
# create 2 new columns, FirstName and LastName by splitting the Name column
df[["height cm'","weight kg'"]] = df['measurements'].str.split(',', expand=True)
# drop the 'Name' column
df.drop('measurements', axis=1, inplace=True)
df.head()
```

```python
df["height cm'"] = df["height cm'"].map(lambda x: x.lstrip("{'Height': '").rstrip("cm'"))
df.head()
```

```python
df["weight kg'"] = df["weight kg'"].map(lambda x: x.lstrip("'Weight': '").rstrip("kg'}"))
df.head()
```

```python
df["height cm'"] = df["height cm'"].astype('float')
df["weight kg'"] = df["weight kg'"].astype('float')
```

```python
df.info()
```

## Superhero Powers

```python
# Load user data
df1 = pd.read_csv(r"C:\Users\paulm\Documents\Saved Data for DataFrames - Coding Dojo\superhero_powers - superhero_powers.csv")
df1.head()
```

```python
df1.info()
```

```python
df1.duplicated().sum()
```

```python
df1.columns = df1.columns.str.lower()
df1.columns
```

```python
df1.describe(include='all')
```

```python
df1['powers'].value_counts()
```

```python
# Create a new column where the single quotes are replaced by double quotes
df1['powers_split'] = df1['powers'].str.split(",")
df1.head()
```

```python
## exploding the column of lists
explode = df1.explode('powers_split')
```

```python
col_to_make = explode['powers_split'].dropna().unique()
```

```python
for col in col_to_make:
    df1[col] = df1['powers'].str.contains(col)
df1
```

```python
# drop the 'Name' column
df1.drop(['powers','powers_split'], axis=1, inplace=True)
df1.head()
```

```python
df1.info()
```

```python
# drop the 'Name' column
combined = pd.merge(df, df1, left_on = 'hero', right_on = 'hero_names')
combined.head()
```

```python
# Super Speed by weight
combined.groupby('Super Speed')["weight kg'"].mean()
```

```python
# Average superhero height based on publisher
combined.groupby('publisher')["height cm'"].mean()
```
